```python
import pandas as pd
df = pd.read_csv("/opt/ml/data/Dataset.csv")
df.to_csv("/opt/ml/data/Dataset.csv",sep = ',',index=False)
df = pd.read_csv("/opt/ml/data/Dataset.csv")
df1 = df.drop("Unnamed: 0",axis=1)
df1
```

```python
# 빅쿼리 생성 
from google.oauth2 import service_account
import pandas_gbq

...
cd = service_account.Credentials.from_service_account_file(
    'key.json',
)
project_id = 'testchatbotquery'
destination_table = 'test2.test_table'
df1.to_gbq(destination_table,project_id,if_exists='replace',credentials=cd) 
print('migration complete')
```

```python
# 빅쿼리 탐색 
from google.cloud import bigquery
client = bigquery.Client (project = project_id , credentials = cd) 
query = "SELECT *  FROM `testchatbotquery.test2.test_table`"

df_result = client.query (query) .to_dataframe ()
```
